```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import pandas as pd
import time
from datetime import datetime
```

```python
url = "https://simplyhired.com/search?l=New+York+State&sr=3" #radius=3
url2 = "https://simplyhired.com/search?q=&l=10006" #zipcode=10006
url3 = "https://simplyhired.com/search?l=New+York+State&sr=0&t=3" #posted day=3
driver = webdriver.Chrome()
driver2 = webdriver.Chrome()
driver3 = webdriver.Chrome()
driver.get(url)
driver2.get(url)
driver3.get(url)
```

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
soup2 = BeautifulSoup(driver2.page_source, 'html.parser')
soup3 = BeautifulSoup(driver3.page_source, 'html.parser')
```

```python
#job_data=[]

jobkey=soup.find_all('div',{'class':'css-f8dtpc'})

date = soup.find_all("p",{"class":"chakra-text css-5yilgw"})
job_title=soup.find_all('div',{'data-testid':'searchSerpJob'})
company_name=soup.find_all('span',{'data-testid':'companyName'})
salary=soup.find_all('p',{'data-testid':'searchSerpJobSalaryEst'})
location=soup.find_all('span',{'data-testid':'searchSerpJobLocation'})
#desc=soup.find_all('div',{'class':'css-1ebprri'})
job_url=soup.find_all('a',{'class':'chakra-button css-1y7j4hn'})


job_page_url=[]
for i in  job_url:
    job_page_url.append(i.get('href'))

jk=[]
for i in jobkey:
    jk.append(i.get('data-jobkey'))


```

```python
liste = list()


for i in range(len(date)):
    date[i] = (date[i].text).strip()
    job_title[i] = (job_title[i].text).strip()
    company_name[i] = (company_name[i].text).strip()
    salary[i] = (salary[i].text).strip()
    location[i] = (location[i].text).strip()
    jk[i]=(jk[i]).strip()
    #full_job_desc[i] = (full_job_desc[i].text).strip()
    job_page_url[i]=(job_page_url[i]).strip()
    scraping_time=datetime.now()
    liste.append([date[i],job_title[i],company_name[i],salary[i],job_page_url[i],location[i],scraping_time,jk[i]])

    df = pd.DataFrame(liste,columns = ['Posting Date','Job Title','Company Name','Salary','Url','Location','Scraping Time','Jobkey'])
    print(df)
```
